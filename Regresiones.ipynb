```python
import pandas as pd 
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#cargamos un dataset de un restaurante
propinas = sns.load_dataset('tips')
propinas.head(10)
```

```python
#Esto compara las variables, la factura total con la propina
sns.lmplot('total_bill','tip',propinas)
```

```python
#as cambianos los colores
sns.lmplot('total_bill','tip',propinas, scatter_kws={'marker':'o','color':'green'},line_kws={'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',propinas, fit_reg=False)
```

```python
propinas.head()
```

```python
#si queremos saber el porcentaje, creamos una columna llamada porcentaje que utiliza la regla de 3 para generarlo
propinas['porcentaje']= 100* propinas['tip']/propinas['total_bill']
propinas.head()
```

```python
#generamos una grafica con el porcentaje
sns.lmplot('size','porcentaje', propinas)
```

```python
#Estas son comparativas de datos por sexo y propinas
sns.lmplot('total_bill','porcentaje',propinas,hue='sex', markers=['x','o'])
```

```python
#ahora jugamos con los datos de los das
sns.lmplot('total_bill','porcentaje', propinas, hue='day')
```
