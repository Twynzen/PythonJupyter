```python
#Mapas de calor
```

```python
import pandas as pd
```

```python
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
```

```python
vuelos = sns.load_dataset('flights')
vuelos.head
```

```python
#Los mapas de calor se hacen en una matriz
vuelos = vuelos.pivot('month','year','passengers')
vuelos
```

```python
#el mapa de calor, lo oscuro con menos valor y lo brillante con más calor
sns.heatmap(vuelos)
```

```python
#Ahora se muestran los valores
sns.heatmap(vuelos, annot=True,fmt='d')
```

```python
#Así se byscan los valores de la matriz
vuelos.loc['May'][1956]
```

```python
#podemeos combertir cualquier valor en el valor central para centrar el maá de calor
valor_central = vuelos.loc['May'][1956]
sns.heatmap(vuelos, center=valor_central, annot=True, fmt='d')
```

```python
#así cambiamos la barra de calor a una forma horizontal
sns.heatmap(vuelos, center=valor_central, annot=True, fmt='d', cbar_kws={'orientation':'horizontal'})
```
