```python
# Combinando estilos
import pandas as pd 
import numpy as np
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
datos= np.random.randn(100)
datos
```

```python
sns.distplot(datos, color="black")
```

```python
#así sin las barras
sns.distplot(datos, color="green", rug=False, hist=False)
```

```python
#Creamos un display de un histograma tanto la curva como las barras, 
#con 25 barras y los parametros de la curva y el histograma
#tendran las configuariciones en una lista exterior al display oroginal
argumentos_curva = {'color':'black', 'label':'Curva'}
argumentos_histograma = {'color':'red', 'label':'Histograma'}
sns.distplot(datos, bins = 25, kde_kws=argumentos_curva,hist_kws=argumentos_histograma)
```

```python
serie = pd.Series(datos)
serie
```
